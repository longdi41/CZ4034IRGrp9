```python
import pandas as pd
import snscrape.modules.twitter as sntwitter
```

```python
#Define search parameters
search_list = ['Russia', 'Ukraine', 'NATO']

num_of_tweets = 100

since_date = '2022-02-24'

date_list = ['2022-02-25','2022-02-26','2022-02-27','2022-02-28','2022-03-01','2022-03-02','2022-03-03','2022-03-04',
             '2022-03-05','2022-03-06','2022-03-07','2022-03-08','2022-03-09','2022-03-10','2022-03-11','2022-03-12',
             '2022-03-13','2022-03-14','2022-03-15','2022-03-16']

language = 'en'

city_list = ['Kyiv', 'Kharkiv', 'Moscow', 'Saint Petersburg', 'Minsk', 'Tokyo', 'Hong Kong', 'Seoul', 'Taipei', 'Singapore', 
             'Berlin', 'Paris', 'London', 'Rome', 'Madrid', 'Bucharest', 'Stockholm', 'Helsinki', 'Vienna', 'Athens', 
             'Prague', 'Lisbon', 'Canberra', 'Tehran', 'Ankara', 'Riyadh']
```

```python
# Creating list to append tweet data to
tweets = []

# Using TwitterSearchScraper to scrape data and append tweets to list
for search in search_list:
    for date in date_list:
        for city in city_list:
            for count, tweet in enumerate(sntwitter.TwitterSearchScraper(f'{search} lang:{language} since:{since_date} until:{date} near:{city}').get_items()):
                if count>num_of_tweets-1:
                    break
                tweets.append([tweet.date, tweet.id, tweet.content, tweet.user.username, tweet.hashtags, 
                           tweet.likeCount, tweet.replyCount, tweet.retweetCount, search, city, tweet.coordinates])
 
```

```python
# Creating a dataframe from the tweets list above
crawled_tweet = pd.DataFrame(tweets, columns=['Datetime', 'Tweet Id', 'Text', 'Username', 'Hashtags', 'Like(s)',
                                           'Reply(s)', 'Retweet(s)','Search term', 'City', 'Coordinates'])
```
